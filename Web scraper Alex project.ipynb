```python
# IMPORT MODULES FOR WEB SCRAPING


import requests
import bs4
import re   # dont know why i imported this
```

```python
req = requests.get('https://en.m.wikipedia.org/wiki/Wikipedia:Multiyear_ranking_of_most_viewed_pages')
soup = bs4.BeautifulSoup(req.text, 'lxml')
```

```python
links = []   # start off with an empty list
```

```python
# This appends all the links on the page to the list of links. TOTAL : 967 links

for link in soup.find_all('a',href = True):
    if link['href'].startswith('#'):
        pass
    elif link['href'].startswith('/'):
        links.append('https://en.wikipedia.org'+link['href'])
    else:
        links.append(link['href'])
```

```python
del links[5]  # found this index number to have a faulty link
links   # just to see all the links
```

```python
file = open('data.txt','a')  # open the .txt file in the same folder in append mode
```

```python
# WHEW this is going to take ages to finish, so please dont re run it lol

for link in links:
    page = requests.get(link)
    text_soup = bs4.BeautifulSoup(page.text, 'lxml')
        
    for item in text_soup.select('p'):      # this creates a list of all items with <p> tags
        
        file.write('\n'+str(links.index(link))+'\n')  # writes index of link, so we can find and remove faulty links 
        file.write(item.text)   # writes all text from link
        
        
file.close()   # closes the file after it's done
```
