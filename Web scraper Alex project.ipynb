```python
import requests
import bs4
import re
```

```python
req = requests.get('https://en.m.wikipedia.org/wiki/Wikipedia:Multiyear_ranking_of_most_viewed_pages')
```

```python
soup = bs4.BeautifulSoup(req.text, 'lxml')
```

```python
links = []
```

```python
for link in soup.find_all('a',href = True):
    if link['href'].startswith('#'):
        pass
    elif link['href'].startswith('/'):
        print('https://en.wikipedia.org'+link['href'])
        links.append('https://en.wikipedia.org'+link['href'])
    else:
        print(link['href'])
        links.append(link['href'])
```
