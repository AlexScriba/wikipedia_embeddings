```python
import requests
import bs4
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Lists_of_automobile-related_articles')
soup = bs4.BeautifulSoup(page.text,'lxml')
```

```python
organised = soup.select('a', href = True)
```

```python
car_links = []
```

```python
for link in soup.select('a', href = True):
    try:
        if link['href'].startswith('#'):
            pass
        elif link['href'].startswith('//'):
            car_links.append('https:'+link['href'])
        elif link['href'].startswith('/'):
            car_links.append('https://en.wikipedia.org/'+link['href'])
        else:
            car_links.append(link['href'])
    except KeyError:
        pass
```

```python

car_links
```

```python
data = open('small_data.txt','a')
```

```python
for i,link in enumerate(car_links):
    if i > 1 and i < 367 and i not in [6,66,69,75,81,86,90,99,156,192,193,199,205,210,217,352]:
        page = requests.get(link)
        text_soup = bs4.BeautifulSoup(page.text, 'lxml')
    
        print(i)
    
        for item in text_soup.select('p'):      # this creates a list of all items with <p> tags
        
            data.write('\n'+str(i)+'\n')  # writes index of link, so we can find and remove faulty links 
            data.write(item.text)   # writes all text from link
        
        
data.close()   # closes the file after it's done
```
